```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = f"../UWA-Project/Output/happiness_merged.csv"
```

```python
happy_data = pd.read_csv(file)
happy_data = happy_data.rename(columns = {"Trust_(Government_Corruption)" : "Trust_(Government_Corruption)_2019"}) 
happy_data
```

```python
happy_data["Country"].nunique()
```

```python
happy_data.describe()
```

```python
# check what is the lowest score per year
max_2015 = happy_data["Happiness_Rank_2015"].max()
max_2016 = happy_data["Happiness_Rank_2016"].max()
max_2017 = happy_data["Happiness_Rank_2017"].max()
max_2018 = happy_data["Happiness_Rank_2018"].max()
max_2019 = happy_data["Happiness_Rank_2019"].max()


print(f"2015 max rank is: {max_2015}\n2016 max rank is: {max_2016}\n2017 max rank is: {max_2017}\
      \n2018 max rank is: {max_2018}\n2019 max rank is: {max_2019}")
```

```python
# Top 10 in Happiness_Rank_2015 with zero values so we cannot use this
happy_data.nsmallest(10, ["Happiness_Rank_2015"])
```

```python
# Top 10 in Happiness_Rank_2015
top_10_2015 = happy_data.loc[(happy_data["Happiness_Rank_2015"] <= 10) & (happy_data["Happiness_Rank_2015"] !=0), :]\
            .sort_values("Happiness_Rank_2015").reset_index(drop = True)
top_10_2015 = top_10_2015[["Country", "Happiness_Rank_2015", "Happiness_Score_2015", "Economy_(GDP_per_Capita)_2015", \
                           "Family_2015", "Health_(Life_Expectancy)_2015", "Freedom_2015", \
                           "Trust_(Government_Corruption)_2015", "Generosity_2015", "Dystopia_Residual_2015"]]
top_10_2015
```

```python
# Retrieve the data per top country
top_country_2015 = []
gdp_2015 = []
family_2015 = []
health_2015 = []
freedom_2015 = []
trust_2015 = []
generosity_2015 = []
dystopia_2015 = []

factors_2015 = [gdp_2015, family_2015, health_2015, freedom_2015, trust_2015, generosity_2015, dystopia_2015]

for country in top_10_2015["Country"]:
    top_country_2015.append(country)
    
for gdp in top_10_2015["Economy_(GDP_per_Capita)_2015"]:
    gdp_2015.append(gdp)

for family in top_10_2015["Family_2015"]:
    family_2015.append(family)
    
for health in top_10_2015["Health_(Life_Expectancy)_2015"]:
    health_2015.append(health)
    
for freedom in top_10_2015["Freedom_2015"]:
    freedom_2015.append(freedom)
    
for trust in top_10_2015["Trust_(Government_Corruption)_2015"]:
    trust_2015.append(trust)
    
for generosity in top_10_2015["Generosity_2015"]:
    generosity_2015.append(generosity)
        
for dystopia in top_10_2015["Dystopia_Residual_2015"]:
    dystopia_2015.append(dystopia)
```

```python
# Horizontal Bar for 2015 top countries
plt.barh(top_country_2015, factors_2015[0], left = np.sum(factors_2015[:0], axis = 0), label = "GDP")
plt.barh(top_country_2015, factors_2015[1], left = np.sum(factors_2015[:1], axis = 0), label = "Family")
plt.barh(top_country_2015, factors_2015[2], left = np.sum(factors_2015[:2], axis = 0), label = "Health")
plt.barh(top_country_2015, factors_2015[3], left = np.sum(factors_2015[:3], axis = 0), label = "Freedom")
plt.barh(top_country_2015, factors_2015[4], left = np.sum(factors_2015[:4], axis = 0), label = "Trust")
plt.barh(top_country_2015, factors_2015[5], left = np.sum(factors_2015[:5], axis = 0), label = "Generosity")
plt.barh(top_country_2015, factors_2015[6], left = np.sum(factors_2015[:6], axis = 0), label = "Dystopia")

plt.legend(loc = (1.05, 0.4))
plt.gca().invert_yaxis()
plt.show()
```

```python
# check what is the lowest score
max_2015 = happy_data["Happiness_Rank_2015"].max()
print(f"2015 max rank is: {max_2015}")
```

```python
# Bottom 10 in Happiness_Rank_2015
bottom_10_2015 = happy_data.nlargest(10, ["Happiness_Rank_2015"]).reset_index(drop = True)\
            .sort_values("Happiness_Rank_2015").reset_index(drop = True)
bottom_10_2015 = bottom_10_2015[["Country", "Happiness_Rank_2015", "Happiness_Score_2015", "Economy_(GDP_per_Capita)_2015",\
                           "Family_2015", "Health_(Life_Expectancy)_2015", "Freedom_2015", \
                           "Trust_(Government_Corruption)_2015", "Generosity_2015", "Dystopia_Residual_2015"]]
bottom_10_2015
```

```python
# Horizontal Bar for 2015 bottom countries
plt.barh(bottom_country_2015, b_factors_2015[0], left = np.sum(b_factors_2015[:0], axis = 0), label = "GDP")
plt.barh(bottom_country_2015, b_factors_2015[1], left = np.sum(b_factors_2015[:1], axis = 0), label = "Family")
plt.barh(bottom_country_2015, b_factors_2015[2], left = np.sum(b_factors_2015[:2], axis = 0), label = "Health")
plt.barh(bottom_country_2015, b_factors_2015[3], left = np.sum(b_factors_2015[:3], axis = 0), label = "Freedom")
plt.barh(bottom_country_2015, b_factors_2015[4], left = np.sum(b_factors_2015[:4], axis = 0), label = "Trust")
plt.barh(bottom_country_2015, b_factors_2015[5], left = np.sum(b_factors_2015[:5], axis = 0), label = "Generosity")
plt.barh(bottom_country_2015, b_factors_2015[6], left = np.sum(b_factors_2015[:6], axis = 0), label = "Dystopia")

plt.legend(loc = (1.05, 0.4))
plt.gca().invert_yaxis()
plt.show()
```
